F.  Desarrolle un subplot con dos gráficos, el primero es un boxplot relacione artist_count (eje x) con streams, y el segundo un histograma de los años de 
lanzamiento. 

```python

# Filtrar solo jugadores fallecidos
df_fallecidos = df[df["dead"] == 1]

# Separar las edades de vida por grupo
vida_diestros = df_fallecidos[df_fallecidos["left"] == "right"]["life"]
vida_zurdos = df_fallecidos[df_fallecidos["left"] == "left"]["life"]

# Crear boxplot
plt.figure(figsize=(6,5))
plt.boxplot([vida_diestros, vida_zurdos],
            labels=["Right (diestro)", "Left (zurdo)"],
            patch_artist=True,
            boxprops=dict(facecolor="skyblue"),
            medianprops=dict(color="red", linewidth=2))

plt.title("Distribución de la edad de vida según mano de bateo")
plt.ylabel("Edad de vida (años)")
plt.show()

# Resumen estadístico
print("Resumen estadístico:")
print("Diestros:\n", vida_diestros.describe(), "\n")
print("Zurdos:\n", vida_zurdos.describe())
```

```python

# Filtrar solo artistas y los streams 
artistas = df[df["artist_count"]=='released_year']
streams = df[df['streams']=='released_year']

# Crear boxplot
plt.figure(figsize=(6,2))
plt.boxplot([artistas, streams],
            labels=["artistas mundiales", "streams"],
            patch_artist=True,
            boxprops=dict(facecolor="skyblue"),
            medianprops=dict(color="red", linewidth=2))

plt.title("")
plt.ylabel("streams")
plt.show()
```
